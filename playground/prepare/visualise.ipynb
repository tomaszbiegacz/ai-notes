```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import shared.prepare as sp
```

```python
housing = sp.loadData_csv_pandas("000_housing.csv")
housing_labels = housing["median_house_value"]
```

```python
cluster_simil = sp.ClusterSimilarity(n_clusters=10, random_state=42)
similarities = cluster_simil.fit_transform(housing[["latitude", "longitude"]], sample_weight=housing_labels)
```

```python
housing.plot(
  kind="scatter", x="longitude", y="latitude", grid=True,
  s=housing["population"]/100, label="population",  # point radius
  c="median_house_value", cmap="jet", colorbar=True, # color range
  legend=True)

plt.plot(cluster_simil.kmeans_.cluster_centers_[:, 1],
         cluster_simil.kmeans_.cluster_centers_[:, 0],
         linestyle="", color="black", marker="X", markersize=10,
         label="Cluster centers")
plt.legend(loc="upper right")
plt.show()
```
