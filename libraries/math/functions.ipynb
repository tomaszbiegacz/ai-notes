```python
import numpy as np
import matplotlib.pyplot as plt

# global variables
figsize=(8, 3)
resolution=1000

def plt_f1(min, max, f, f_name=None):
  if f_name is None:
    f_name = f.__name__
  x = np.linspace(min, max, resolution)
  fig, ax = plt.subplots(figsize=figsize)
  ax.spines['left'].set_position('zero')
  ax.spines['bottom'].set_position('zero')
  ax.spines['right'].set_color('none')
  ax.spines['top'].set_color('none')

  ax.plot(x, f(x), label='f');
  ax.set_title(f_name)
  xticks = np.arange(np.floor(min), np.ceil(max), 1).astype(int)
  ax.set_xticks(np.extract(xticks != 0, xticks))
  ax.legend(loc='lower right');
```

```python
plt_f1(0.001, 10, np.log)
```

```python
plt_f1(-10, 10, np.arctan)
```

```python
plt_f1(-1.5, 1.5, np.tan)
```

```python
plt_f1(-10, 10, np.cbrt)
```

```python
plt_f1(-10, 10, lambda v: v, 'x')
```

```python
plt_f1(0, 10, lambda v: np.power(v, 0.7), "^0.7")
```

```python
plt_f1(0, 10, np.sqrt)
```

```python
plt_f1(-10, 10, np.square)
```
