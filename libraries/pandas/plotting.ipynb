```python
from pathlib import Path
from pandas.plotting import scatter_matrix

import pandas as pd
import numpy as np
```

```python
housing = pd.read_csv(Path("work/data/000_housing.csv"))
```

## Bar

```python
distr_bins_2 = [0, 1.5, 3.0, 4.5, 6.0, np.inf]
distr_data_2 = pd.cut(housing["median_income"], bins=distr_bins_2, labels=np.arange(1, len(distr_bins_2), 1))
distr_data_2.value_counts().sort_index().plot.bar(figsize=(4, 4))
```

## Cumsum

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts.cumsum().plot(figsize=(4, 4));
```

## Histogram

```python
housing.hist(column="housing_median_age", figsize=(3,3))
```

## Correlation

```python
scatter_matrix(housing[["median_house_value", "median_income"]], figsize=(4, 4))
```

## Geographic data

```python
housing.plot(
  kind="scatter", x="longitude", y="latitude", grid=True,
  alpha=0.2, figsize=(3,3)
  )
```

```python
housing.plot(
  kind="scatter", x="longitude", y="latitude", grid=True,
  s=housing["population"]/100, label="population",  # point radius
  c="median_house_value", cmap="jet", colorbar=True, # color range
  legend=True)
```
